# Importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Input Data (File I/O)

```python
df = pd.read_csv("inputData.csv")
```

# Data preparation

```python
df['Player Power'] = df['playerOVR'] + df['playerCHEM']/10
df['Opponent Power'] = df['opponentOVR'] + df['opponentCHEM']/10
df.sort_values("Player Power", ascending=False)
```

# Calculate the goal difference and color matching

```python
df['Power Diff'] = (df['Player Power'] - df['Opponent Power']).apply(np.ceil).astype(np.int)
df
```

### Matching color with a if-elif-else function

```python
def convertColor(row):
    if row['Power Diff'] >= 10:
        col = "#24693D"
    elif row['Power Diff'] == 9:
        col = "#27763D"
    elif row['Power Diff'] == 8:
        col = "#2F8143"
    elif row['Power Diff'] == 7:
        col = "#3F8D4B"
    elif row['Power Diff'] == 6:
        col = "#4E9951"
    elif row['Power Diff'] == 5:
        col = "#5DA656"
    elif row['Power Diff'] == 4:
        col = "#6BB260"
    elif row['Power Diff'] == 3:
        col = "#78BF6C"
    elif row['Power Diff'] == 2:
        col = "#85CA77"
    elif row['Power Diff'] == 1:
        col = "#98D687"
    elif row['Power Diff'] == 0:
        col = "#CED7C3"
    elif row['Power Diff'] == -1:
        col = "#FDAA9A"
    elif row['Power Diff'] == -2:
        col = "#FB9985"
    elif row['Power Diff'] == -3:
        col = "#F98770"
    elif row['Power Diff'] == -4:
        col = "#F5745F"
    elif row['Power Diff'] == -5:
        col = "#F16051"
    elif row['Power Diff'] == -6:
        col = "#EA4C47"
    elif row['Power Diff'] == -7:
        col = "#DF3927"
    elif row['Power Diff'] == -8:
        col = "#D0283D"
    elif row['Power Diff'] == -9:
        col = "#C01A3B"
    else:
        col = "#AE123A"
    return col
```

### Add color to corresponding match

```python
df['Color'] = df.apply(convertColor, axis=1)
df
```

```python
home = df[df['playerLeague']==0]
away = df[df['playerLeague']==1]
```

# Sorting the Player by Power

```python
home = home.sort_values(by=['Player Power'], axis=0, ascending=False)
away = away.sort_values(by=['Player Power'], axis=0, ascending=False)
```

# Output the required code for plotting Team Sankey Diagram

### Home Team 

```python
f = open("home_vis.txt", "w+", encoding="utf-8")
for index, row in home.iterrows():
    abc = (row['playerName'] + " (" + str(row['playerOVR']) + " " + str(row['playerCHEM']) + "):, " + row['opponentName'] + " (" + str(row['opponentOVR']) + " " + str(row['opponentCHEM']) + "), " + str(row['goal scored']) + "," + " " + row['Color'])
    f.write(abc + "\n")
f.close()
```

### Away Team

```python
f = open("away_vis.txt", "w+", encoding="utf-8")
for index, row in away.iterrows():
    abc = (row['playerName'] + " (" + str(row['playerOVR']) + " " + str(row['playerCHEM']) + "):, " + row['opponentName'] + " (" + str(row['opponentOVR']) + " " + str(row['opponentCHEM']) + "), " + str(row['goal scored']) + ", " + row['Color'])
    f.write(abc + "\n")
f.close()
```
